```python
import seaborn as sns

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic_data.head()
```

```python
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a barplot of the number of survivors by class
plt.figure(figsize=(10, 6))
sns.barplot(x='class', y='survived', data=titanic_data, ci=None, palette='muted')

# Add title and labels
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')

# Show the plot
plt.show()
```

```python
sns.barplot(titanic_data, x='pclass', y='survived', hue=True)
plt.title('difference bw these things')
plt.xlabel('this is the part of fare and it will tell you the numbers in it')
plt.ylabel('this is the part of age and that will show you the age of all the people')
plt.show()
```
