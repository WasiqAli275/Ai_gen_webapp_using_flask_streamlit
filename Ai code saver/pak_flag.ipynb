```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def create_pakistan_flag():
    # Set up the figure and axis
    fig, ax = plt.subplots(figsize=(9, 6))
    
    # Flag dimensions (3:2 ratio)
    flag_width = 9
    flag_height = 6
    white_stripe_width = flag_width / 4
    
    # Create the green field
    green_rect = patches.Rectangle((white_stripe_width, 0), 
                                 flag_width - white_stripe_width, 
                                 flag_height, 
                                 facecolor='#01411C',  # Dark green
                                 edgecolor='none')
    ax.add_patch(green_rect)
    
    # Create the white vertical stripe
    white_rect = patches.Rectangle((0, 0), 
                                 white_stripe_width, 
                                 flag_height, 
                                 facecolor='white',
                                 edgecolor='none')
    ax.add_patch(white_rect)
    
    # Draw the crescent moon
    # Center position for the crescent and star
    center_x = white_stripe_width + (flag_width - white_stripe_width) * 0.6
    center_y = flag_height * 0.5
    
    # Outer circle of crescent
    outer_circle = patches.Circle((center_x, center_y), 
                                radius=0.8, 
                                facecolor='white',
                                edgecolor='none')
    ax.add_patch(outer_circle)
    
    # Inner circle (to create crescent shape)
    inner_circle = patches.Circle((center_x + 0.25, center_y), 
                                radius=0.7, 
                                facecolor='#01411C',  # Same as green field
                                edgecolor='none')
    ax.add_patch(inner_circle)
    
    # Draw the five-pointed star
    # Star position (to the right of crescent)
    star_x = center_x + 1.2
    star_y = center_y + 0.2
    
    # Create star using polygon
    # Five-pointed star coordinates
    star_radius = 0.3
    angles = np.linspace(0, 2*np.pi, 11)  # 11 points for 5-pointed star
    star_x_coords = []
    star_y_coords = []
    
    for i, angle in enumerate(angles):
        if i % 2 == 0:
            # Outer points
            r = star_radius
        else:
            # Inner points
            r = star_radius * 0.4
        
        x = star_x + r * np.cos(angle - np.pi/2)
        y = star_y + r * np.sin(angle - np.pi/2)
        star_x_coords.append(x)
        star_y_coords.append(y)
    
    star = patches.Polygon(list(zip(star_x_coords, star_y_coords)),
                         facecolor='white',
                         edgecolor='none')
    ax.add_patch(star)
    
    # Set axis limits and remove axes
    ax.set_xlim(0, flag_width)
    ax.set_ylim(0, flag_height)
    ax.set_aspect('equal')
    ax.axis('off')
    
    # Add title
    plt.title('Flag of Pakistan', fontsize=16, pad=20)
    
    # Show the flag
    plt.tight_layout()
    plt.show()

# Create the flag
create_pakistan_flag()
```
