```python
# For interactive charts (if using locally)
import plotly.express as px
```

```python
# ## üìÇ Step 2: Load the Dataset
df = pd.read_csv('anime.csv')
```

```python
# Display first few rows
df.head()
```

```python

# ## üìà Step 3: Basic Info and Missing Values
print("\nDataset Info:")
print(df.info())
```

```python

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ## üîç Step 4: Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())
```


### üé® Step 5: **Data Visualization**

```python
# ‚≠ê Rating Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['rating'], bins=20, kde=True, color='skyblue')
plt.title('Rating Distribution of Anime')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python

# üìä Type of Anime Count
plt.figure(figsize=(8,5))
sns.countplot(y='type', data=df, order=df['type'].value_counts().index, palette='pastel')
plt.title('Number of Anime by Type')
plt.xlabel('Count')
plt.ylabel('Type')
plt.show()
```

```python
# üèÖ Top 10 Most Popular Anime
top_popular = df.sort_values('members', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='members', y='name', data=top_popular, palette='muted')
plt.title('Top 10 Most Popular Anime by Members')
plt.xlabel('Number of Members')
plt.ylabel('Anime Name')
plt.show()
```

```python
# üåü Top Rated Anime (with more than 1000 members)
top_rated = df[df['members']>1000].sort_values('rating', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='rating', y='name', data=top_rated, palette='viridis')
plt.title('Top 10 Highest Rated Anime (with >1000 members)')
plt.xlabel('Rating')
plt.ylabel('Anime Name')
plt.show()
```

```python
# üìà Scatter Plot: Members vs Rating
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='members', y='rating', hue='type', alpha=0.7)
plt.title('Members vs Rating by Type')
plt.xlabel('Number of Members')
plt.ylabel('Rating')
plt.show()
```

```python
# ## üìä Interactive Plot with Plotly (Optional)
fig = px.scatter(df, x='members', y='rating', color='type', hover_data=['name'],
                 title='Interactive Members vs Rating by Type')
fig.show()
```
